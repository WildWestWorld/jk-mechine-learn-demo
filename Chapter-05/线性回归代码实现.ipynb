## 一元线性回归

```python
import numpy as np 
from sklearn import datasets
import matplotlib.pyplot as plt
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
boston = datasets.load_boston()
```

```python
#获取波士顿放假的 平均房间数 ,5就是 平均房间数
x= boston.data[:,5]
y= boston.target
#y轴的最大值是50,所以我们手段筛选 数据为小于50的数据
x= x[y<50]
y=y[y<50]

plt.scatter(x,y)
plt.show()
```

```python
#划分数据集
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.3,random_state=0)

plt.scatter(x_train, y_train)
plt.show()
```

## 一元线性回归公式实现

  
使用最小二乘法，就能找到最小的k和b  x^ ,y^=x,y均值

```python
def fit(x,y):
    #k的分子
    k_up = np.sum( (x - np.mean(x)) * (y - np.mean(y)) )
    k_bottom = np.sum((x - np.mean(x))**2 )
    k = k_up / k_bottom

    b = np.mean(y) - k * np.mean(x)
    return k,b
```

```python
k,b = fit(x_train,y_train)
k,b
```

```python
plt.scatter(x_train,y_train)
plt.plot(x_train, k*x_train +b,c='r')
plt.show()
```

```python
#测试 测试集
plt.scatter(x_test,y_test)
plt.plot(x_test, k*x_test +b,c='r')
plt.show()
```

## sklearn 实现一元线性回归

```python
#导入线性回归模型
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()

lin_reg.fit(x_train.reshape(-1,1),y_train)
```

```python
y_predict = lin_reg.predict(x_test.reshape(-1,1))
```

```python
plt.scatter(x_test,y_test)
plt.plot(x_test,y_predict,c='r')
plt.show()
```
